```python
# Because TotalCharges is in Object while It is tenure * MonthlyCharges
# So, replacing blanks with 0 as tenure is 0 and no total charges are recorded.
df["TotalCharges"] = df["TotalCharges"].replace(" ","0")
df["TotalCharges"] = df["TotalCharges"].astype("float")
```

```python
def conv_to_boolean(value):
    if value == 0:
        return "no"
    else:
        return "yes"
```

```python
# converted 0 and 1 values in senior citizen to yes and no.
df["SeniorCitizen"] = df["SeniorCitizen"].apply(conv_to_boolean)
```

```python
# Churm Count plotting
ax = sns.countplot(data = df, x = df['Churn'])
ax.bar_label(ax.containers[0])
plt.title("Customer Churn Count Plot")
plt.show()
```

```python
plt.figure(figsize=(3,3))
gb = df.groupby("Churn").agg({'Churn':"count"})
gb
plt.pie(gb['Churn'], labels = gb.index, autopct = "%1.2f%%")
plt.title("Customer Churn %")
plt.show()
```

### Explore Reason behind 26.54% Customer Churn / 1869 customer churn count

```python
plt.figure(figsize = (6,5))
ax = sns.countplot(x = df['gender'], data = df, hue = 'Churn')
for container in ax.containers:
    ax.bar_label(container)
plt.title("Churn by gender")
plt.show()
```

```python
plt.figure(figsize = (4,5))
ax = sns.countplot(x="SeniorCitizen", data=df)
for container in ax.containers:
    ax.bar_label(container)
plt.title("Churn by Senior Citizen")
plt.show()
```

```python
plt.figure(figsize = (5,4))
ax = sns.countplot(x="SeniorCitizen", data=df, hue="Churn")
for container in ax.containers:
    ax.bar_label(container)
plt.title("Churn by Senior Citizen")
plt.show()
```

```python

cross_tab = pd.crosstab(df['SeniorCitizen'], df['Churn'])

cross_tab_percent = cross_tab.div(cross_tab.sum(axis=1), axis=0) * 100

ax = cross_tab_percent.plot(
    kind='bar', 
    stacked=True, 
    figsize=(6,5), 
    colormap='Set2'
)

for container in ax.containers:
    ax.bar_label(container, fmt='%.2f%%', label_type='center')

plt.title('Churn Percentage by Senior Citizen')
plt.ylabel('Percentage')
plt.xlabel('Senior Citizen')
plt.legend(title='Churn')
plt.show()
```

## So, Greater% of Senior Citizen have churn

```python
df.columns.values
```

```python
features = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

# Set up the figure and axes
fig, axes = plt.subplots(5, 2, figsize=(12, 30))  # 3 rows, 2 columns
axes = axes.flatten()  # Flatten into 1D array for easy looping

# Loop through features and create countplots
for i, feature in enumerate(features):
    ax = sns.countplot(x=feature, data=df, ax=axes[i], hue='Churn')
    for container in ax.containers:
        ax.bar_label(container)
    axes[i].set_title(f'Countplot of {feature}', fontsize=14)
    axes[i].set_xlabel('')  # Remove x label to make it cleaner
    axes[i].set_ylabel('Count')

# Adjust layout
plt.tight_layout()
plt.show()
```

### So, Customer who have not choice Techsuppport, Onlinebackup, device protection and online security are likely to Churn more.

### and Customer who are using fiber services are likely to churn more
